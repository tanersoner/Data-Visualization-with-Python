```python
import pandas as pd
```

```python
df = pd.read_csv('ride_sharing_new.csv')
print(df.head())
```

```python
df.info()
```

In Data Set duration data type is object. I want to convert it int. I need to make to strip minutes.

```python
# Strip duration of minutes
df['duration_s'] = df['duration'].str.strip('minutes')
# Convert duration to integer
df['duration_time'] = df['duration_s'].astype('int')
```

```python
# Write an assert statement making sure of operation
assert df['duration_time'].dtype == 'int'
```

```python
# Make control and print new columns. Then calculate mean value of ride duration
print(df[['duration','duration_s','duration_time']])
print(df['duration_time'].dtype)
print(df['duration_time'].mean())
```

I want to find duplicate values. I need to make new dataframe. That is df_duplicated. 

```python
# Find duplicates
df_duplicated = df[df.duplicated()]
#df_duplicated = df.duplicated('bike_id', keep=False)
df_duplicated.info()
```

```python
# There is no duplicated row.
print(df_duplicated)
```

I want to find the unique values of user type, user gender and station names in the data set.

```python
# Print unique values of station name, gender and user type
print('User Type: ', df['user_type'].unique(), '\n')
print('User Gender: ', df['user_gender'].unique(), '\n')
# Find and combine unique values in both columns 'station_A_name' & 'station_B_name'
station_a = df['station_A_name'].unique()
station_b = df['station_B_name'].unique()
unique_combined = set(station_a).union(set(station_b))
print('Number of Station Name: ', len(unique_combined), '\n')
```

I want to encoding for user_type's column. I kontrol how many user types. I might need it for my machine learning

```python
# Convert User type column with one-hot encoding
df_dummies = pd.get_dummies(df, columns=['user_type'])

# Convert only the one-hot encoded columns to integer type
one_hot_columns = [col for col in df_dummies.columns if 'user_type_' in col]
df_dummies[one_hot_columns] = df_dummies[one_hot_columns].astype('int')

print(df_dummies)
```
