```python
import pandas as pd
```

```python
df = pd.read_excel("hessen.xlsx", sheet_name="Bevölkerung", header=0, usecols="B:T", dtype={"Sutun1": str, "Sutun2": float})
```

```python
df.head()
```

```python
df.drop(index=df.index[:3], inplace=True)
df.head()
```

```python
new_column_names = ['Gebiet', 'Fläche', 'Insgesamt','Anzahl_männlich', '%_männlich', 'Anzahl_weiblich', '%_weiblich', 'Anzahl_Abnahme', '%_Abnahme', 'Einwohnerinnen_auf_1000_Einwohner', 'Einwohner/-innen_je_km²','Anzahl_unter_6', '%_unter_6', 'Anzahl_6-15','%_6-15','Anzahl_15-65','%_15-65','Anzahl_65_mehr','%_65_mehr' ]
```

```python
# Assign new names to each column
for i, name in enumerate(new_column_names):
    df.rename(columns={df.columns[i]: name}, inplace=True)

print(df)
```
