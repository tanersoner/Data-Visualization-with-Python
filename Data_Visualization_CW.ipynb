```python
fig,ax=plt.subplots(2,1, figsize=(7,9))
ax[0].bar(labels, age_men, label='Men',  width=0.5)
ax[0].set_title('Age mean by Gender')
ax[0].set_ylabel('Ages')
ax[0].legend()

ax[1].bar(labels, age_women, label='Women',  width=0.5)
ax[1].set_title('Age mean by Gender')
ax[1].set_ylabel('Ages')
ax[1].legend()

for index,data in enumerate(age_women):
    plt.text(x=index , y =data , s=f"{data}" , fontdict=dict(fontsize=15))
plt.tight_layout()
plt.show();
```
