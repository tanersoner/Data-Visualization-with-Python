```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(('sin', 'cos'));
```

```python
fig = plt.figure()

plt.figure(figsize=(11,9))
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), '-')

plt.ylabel('y axis')
plt.title('Sin graphic')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), '--', linewidth=4, color='red')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Cos graphic');
```
